```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#import dataset
dataset = pd.read_csv("spam2.csv")
dataset
```

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
scaled_dataset=scaler.fit_transform(dataset)
scaled_dataset_table=pd.DataFrame(data=scaled_dataset, columns= dataset.columns)
scaled_dataset_table.describe()
```

```python
#Principal componenet analysis

from sklearn.decomposition import PCA
pca = PCA(n_components = 5)
new_data = pca.fit_transform(scaled_dataset_table)
# final dataset to be pridicted
principal_Dataset_table = pd.DataFrame(data = new_data
                                       , columns = ['PC1', 'PC2','PC3','PC4','PC5'])
principal_Dataset_table.head()
```

```python
#defining X and y
X = principal_Dataset_table.iloc[:, :-1] # all attributes except last one
y = dataset.iloc[:, -1] # last column as target variable
```

```python
#training testing data splitting
from sklearn.model_selection import train_test_split
# Split the data considering the testing corpus as 30% and random state to 101.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
#creating decision tree model
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(max_depth=4)
model.fit(X_train,y_train)
DecisionTreeClassifier(max_depth=4)
model.score(X_train, y_train)

model.get_params()
```

```python
#testig data accuracy
from sklearn.metrics import accuracy_score
predictions_test=model.predict(X_test)
print("DecisionTrees Test Accuracy: ",accuracy_score(y_test, predictions_test))
```

```python
#training data accuracy
predictions_train = model.predict(X_train)
print("DecisionTrees Train Accuracy: ", accuracy_score(y_train,predictions_train))
```

```python
#get the image of decision tree
from sklearn import tree
plt.figure(figsize=(100,100))
tree.plot_tree(model,filled=True)
plt.show()
```

```python
#classification report
from sklearn.metrics import confusion_matrix, classification_report

print(classification_report(y_test,predictions_test))
```

```python
#confusion matrix
conmtrx= confusion_matrix(y_test,predictions_test)
conmtrx
```
