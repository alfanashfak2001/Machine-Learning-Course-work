```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the dataset
dataset = pd.read_csv('spam2.csv')
dataset
```

```python
#checking for null values in the dataset
print(dataset.isnull().sum())
```

```python
X=dataset.drop(['class'],axis=1) #removing the labled class column from data set and take other columns
y=dataset['class'] #taking the class column as the result
X
```

```python
y
```

```python
# Splitting the data set to training set and testing set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
X_train, X_test, y_train, y_test
```

```python
# Feature Scaling process
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
X_train,X_test
```

```python
# Fitting classifier to the Training set
from sklearn.neighbors import KNeighborsClassifier
#as usual using k=5
classifier = KNeighborsClassifier(n_neighbors=3, metric='minkowski', p=2)

classifier.fit(X_train,y_train)
```

```python
# Predicting the Test data set results
y_pred = classifier.predict(X_test)
```

```python
#Confusion Matrix
from sklearn.metrics import confusion_matrix,classification_report
con_metrix = confusion_matrix(y_test, y_pred)
con_metrix
```

```python
#print classification report accuracy check
print(classification_report(y_test,y_pred))
```

```python
#to find the best k value
error_rate=[] #creating an array
for i in range(1,40):
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i=knn.predict(X_test)
    #calculating error value
    error_rate.append(np.mean(pred_i != y_test))
```

```python
#plotting the error value with k value to find the elbow point
plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate,color='red',linestyle='dashed',marker='o',markerfacecolor="blue",markersize=10)
plt.title('error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```
